```python
#the number of open complaints range from 1 to 5 so I will calculate the mode and substitute
file1['number of open complaints'].mode()
```

```python
#to make sure this substitution makes sense I will check if 0 is the most common one by a big difference
file1['number of open complaints'].value_counts()
```

```python
#it is by far the mos common one so it makes sense to substitute
file1 = (file1.fillna({'number of open complaints': '0'}))
file1.isna().sum()
```
